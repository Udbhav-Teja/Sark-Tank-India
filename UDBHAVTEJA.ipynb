```python
import os
import pandas as pd
from sklearn import *
os.chdir('E:/python codes/Deep Learning/Shark_Tank')
df=pd.read_csv('Sharkus.csv',encoding = "ISO-8859-1")
df
```

```python
df.columns
```

```python
df.drop(['Season Number', 'Season Start', 'Season End', 'Episode Number',
       'Pitch Number', 'Original Air Date','Business Description', 'Pitchers City','Number of sharks in deal',
       'Pitchers State', 'Pitchers Average Age', 'Entrepreneur Names',
       'Company Website', 'Multiple Entrepreneurs', 'US Viewership','Investment Amount Per Shark',
       'Equity Per Shark', 'Royalty Deal', 'Loan','Original Ask Amount', 'Original Offered Equity', 'Valuation Requested', 'Total Deal Equity', 'Deal Valuation', 'Guest Investment Amount',
       'Guest Investment Equity', 'Guest Name', 'Barbara Corcoran Investment Equity',
       'Mark Cuban Investment Equity', 'Lori Greiner Investment Equity', 'Robert Herjavec Investment Equity',
       'Daymond John Investment Equity', 'Kevin O Leary Investment Equity','Barbara Corcoran Present',
       'Mark Cuban Present', 'Lori Greiner Present', 'Robert Herjavec Present',
       'Daymond John Present', 'Kevin O Leary Present','Startup Name'],axis=1,inplace=True)
```

```python
df.info()
```

```python
def replace(val):
    if val>0:
        val=1
    else:
        val=0
    return val
```

```python

df['Barbara Corcoran Investment Amount'].fillna(0,inplace=True)
df['Mark Cuban Investment Amount'].fillna(0,inplace=True)
df['Lori Greiner Investment Amount'].fillna(0,inplace=True)
df['Robert Herjavec Investment Amount'].fillna(0,inplace=True)
df['Daymond John Investment Amount'].fillna(0,inplace=True)
df['Kevin O Leary Investment Amount'].fillna(0,inplace=True)
#Filling 1'S
df['Barbara Corcoran Investment Amount']=df['Barbara Corcoran Investment Amount'].apply(replace)
df['Mark Cuban Investment Amount'] = df['Mark Cuban Investment Amount'].apply(replace)
df['Lori Greiner Investment Amount'] =df['Lori Greiner Investment Amount'].apply(replace)
df['Robert Herjavec Investment Amount'] =df['Robert Herjavec Investment Amount'].apply(replace)
df['Daymond John Investment Amount'] =df['Daymond John Investment Amount'].apply(replace)
df['Kevin O Leary Investment Amount'] =df['Kevin O Leary Investment Amount'].apply(replace)

df['Total Deal Amount'].fillna(0,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Pitchers Gender']=le.fit_transform(df['Pitchers Gender'])
df['Industry']=le.fit_transform(df['Industry'])
```

```python
df.info()
```

```python
df.to_csv(r"E:\python codes\Deep Learning\Shark_Tank\new.csv")
```

Model Building

```python
x=df.drop(['Barbara Corcoran Investment Amount'],axis=1)
y=df['Barbara Corcoran Investment Amount']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier 
lr=RandomForestClassifier ()
lr.fit(x_train,y_train)
predictions=lr.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score

print(100*accuracy_score(y_test,predictions))
```

```python
from sklearn.linear_model import LogisticRegression
l=LogisticRegression()
l.fit(x_train,y_train)
predictions=l.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score

print(100*accuracy_score(y_test,predictions))
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
dtc.fit(x_train,y_train)
predictions=dtc.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print(100*accuracy_score(y_test,predictions))
```
